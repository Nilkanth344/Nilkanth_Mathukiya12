Encoding data

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
```

```python
df.dtypes == 'object'
```

```python
df['circle'] = le.fit_transform(df['circle'])
df['plan_type'] = le.fit_transform(df['plan_type'])
df['data_pack_type'] = le.fit_transform(df['data_pack_type'])
df['device_type'] = le.fit_transform(df['device_type'])
df['network_usage_type'] = le.fit_transform(df['network_usage_type'])
```

```python
df.head()
```

```python
df['churn_status'].value_counts()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
```

```python
X = df.drop(columns=['churn_status'])
y = df['churn_status']
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,random_state=42,test_size=0.2)
```

```python
lr = LogisticRegression()
```

```python
lr.fit(X_train,y_train)
```

```python
y_pred = lr.predict(X_test)
y_pred
```

```python
from sklearn.metrics import accuracy_score, classification_report
```

```python
accuracy = accuracy_score(y_pred,y_test)*100
accuracy
```

```python
report = classification_report(y_pred,y_test)
report
```
